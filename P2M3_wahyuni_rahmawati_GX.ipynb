```python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'data_clean'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'data_clean'
path_to_data = 'https://raw.githubusercontent.com/wahyunirahmawati/data/main/P2M3_rahma_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

```python
# Expectation 1 : Column `permalink` must be unique

validator.expect_column_values_to_be_unique('permalink')
```

```python
# Expectation 2 : Column `founded_year` range between 1902-2014

validator.expect_column_values_to_be_between(
    column='founded_year', min_value=1902, max_value=2014
)
```

```python
# Expectation 3 : Column `status` must contain one of the following 3 things :
#operting
#acquired
#closed

validator.expect_column_values_to_be_in_set('status', ["operating","acquired","closed"])
```

```python
# Expectation 4 : Column `seed` must in form of integer

validator.expect_column_values_to_be_in_type_list('seed', ['integer'])
```

```python
# Expectation 5 : Column `founded_month` the format year and month
validator.expect_column_values_to_match_strftime_format('founded_month', "%Y-%m")
```

```python
# Expectation 6 :If we want to allow these outliers, we can add some tolerance to the check by using the mostly parameter

validator.expect_column_values_to_be_between('founded_year', min_value=1902, max_value=2014, mostly=0.99)
```

```python
# Expectation 7 :check whether our expectation that the name column is present

validator.expect_column_to_exist('name')
```

```python
# Save into Expectation Suite

validator.save_expectation_suite(discard_failed_expectations=False)
```

```python
# Create a checkpoint

checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

```python
# Run a checkpoint

checkpoint_result = checkpoint_1.run()
```
